```python
# !pip install pandas
# !pip install matplotlib

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

#### Step 1: Load the dataset


```python
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"

df = pd.read_csv(file_path)
```


### Task 1: Exploring Relationships with Scatter Plots



#### 1. Scatter Plot for Age vs. Job Satisfaction



Visualize the relationship between respondents' age (`Age`) and job satisfaction (`JobSatPoints_6`). Use this plot to identify any patterns or trends.




```python
## Write your code here
df['Age'].unique()
```

```python
# Plot
# sns.scatterplot(data=df, x='JobSat_lvl', y='Age_num')
sns.scatterplot(data=df, x='JobSatPoints_6', y='Age')
```

#### 2. Scatter Plot for Compensation vs. Job Satisfaction


Explore the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`) using a scatter plot.


```python
## Write your code here
# Remove outliers using IQR method
Q1 = df['ConvertedCompYearly'].quantile(0.25)
Q3 = df['ConvertedCompYearly'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out outliers
df_clean = df[(df['ConvertedCompYearly'] >= lower_bound) & (df['ConvertedCompYearly'] <= upper_bound)]

# Plot
plt.figure(figsize=(15, 6))
sns.scatterplot(data=df_clean, x='JobSatPoints_6', y='ConvertedCompYearly')
```

### Task 2: Enhancing Scatter Plots


#### 1. Scatter Plot with Trend Line for Age vs. Job Satisfaction



Add a regression line to the scatter plot of Age vs. JobSatPoints_6 to highlight trends in the data.


```python
# mapping job satisfaction .................................
def map_sat(score):
    if score < 30:
        return 'Low'
    elif score < 60:
        return 'Medium'
    elif score < 80:
        return 'High'
    else:
        return 'Very High'

df['JobSat_lvl'] = df['JobSatPoints_6'].apply(map_sat)

# mapping age .............................................................
age_map = {
    'Under 18 years old': 15,
    '35-44 years old': 40,
    '45-54 years old': 50,
    '18-24 years old': 20,
    '25-34 years old': 30,
    '55-64 years old': 60,
    'Prefer not to say': 0,
    '65 years or older': 70
}
df['Age_num'] = df['Age'].map(age_map)
```

```python
## Write your code here
plt.figure(figsize=(9, 6))
sns.regplot(
    data=df,
    x='Age_num', y='JobSatPoints_6',
    ci=95,
    scatter_kws={'alpha': 0.35, 's': 18},
    line_kws={'color': 'crimson', 'lw': 2})
```

#### 2. Scatter Plot for Age vs. Work Experience


Visualize the relationship between Age (`Age`) and Work Experience (`YearsCodePro`) using a scatter plot.


```python
plot_data1 = df.dropna(subset=['YearsCodePro']).reset_index()

# Define mapping function
def map_years_code(x):
    if x == 'Less than 1 year':
        return 0.5
    elif x == 'More than 50 years':
        return 51
    else:
        try:
            return float(x)
        except:
            return np.nan  # handles any unexpected non-numeric strings

# Apply to column
plot_data1['YearsCodePro_num'] = plot_data1['YearsCodePro'].apply(map_years_code)
```

```python
## Write your code here

plt.figure(figsize=(9, 6))
sns.regplot(
    data=plot_data1,
    x='Age_num', y='YearsCodePro_num',
    ci=95,
    scatter_kws={'alpha': 0.35, 's': 18},
    line_kws={'color': 'crimson', 'lw': 2})
```

### Task 3: Combining Scatter Plots with Additional Features


#### 1. Bubble Plot of Compensation vs. Job Satisfaction with Age as Bubble Size



Create a bubble plot to explore the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`), with bubble size representing age.


```python
## Write your code here
plt.figure(figsize=(15,9))
sns.scatterplot(data=df_clean, x='JobSatPoints_6', y='ConvertedCompYearly', hue='Age', size='Age', sizes=(20, 300), alpha=0.5, palette='magma')
plt.xlabel('Job Satisfaction')
plt.ylabel('Compensation')
plt.title('relationship between yearly compensation Compensation and job satisfaction Job Satisfaction')
```

#### 2. Scatter Plot for Popular Programming Languages by Job Satisfaction


Visualize the popularity of programming languages (`LanguageHaveWorkedWith`) against job satisfaction using a scatter plot. Use points to represent satisfaction levels for each language.


```python
# Write your code here
lang_cols = ['LanguageHaveWorkedWith', 'JobSatPoints_6']
lang = df[lang_cols].copy()

lang = lang.dropna(subset=['LanguageHaveWorkedWith', 'JobSatPoints_6'])

# Explode semicolon-separated languages
lang['Language'] = lang['LanguageHaveWorkedWith'].str.split(';')
lang_ex = lang.explode('Language')
lang_ex['Language'] = lang_ex['Language'].str.strip()
lang_ex = lang_ex[lang_ex['Language'] != '']
```

```python
top_n = 15  # adjust for readability
counts = lang_ex['Language'].value_counts()
top_langs = counts.head(top_n).index

plot = lang_ex[lang_ex['Language'].isin(top_langs)].copy()
plot['Language'] = pd.Categorical(plot['Language'], categories=top_langs, ordered=True)
```

```python
sns.set(style='whitegrid')

plt.figure(figsize=(14, 7))
# Points: each respondents satisfaction
sns.stripplot(
    data=plot,
    x='Language', y='JobSatPoints_6',
    jitter=0.3, alpha=0.35, size=3, color='#1f77b4'
)

# Overlay medians for each language
medians = plot.groupby('Language')['JobSatPoints_6'].median().reindex(top_langs)
sns.pointplot(
    x=medians.index, y=medians.values,
    color='crimson', join=False, markers='D', scale=1, errwidth=0
)

plt.ylim(0, 100)
plt.xticks(rotation=40, ha='right')
plt.xlabel(f'Programming language (Top {top_n} by respondents)')
plt.ylabel('Job satisfaction (0100)')
plt.title('Programming languages vs job satisfaction (points = respondents; diamonds = median)')
plt.tight_layout()
plt.show()
```

### Task 4: Scatter Plot Comparisons Across Groups


#### 1. Scatter Plot for Compensation vs. Job Satisfaction by Employment Type


Visualize the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`), categorized by employment type (`Employment`). Use color coding or markers to differentiate between employment types.


```python
## Write your code here

# Keep top-k employment categories for clearer legend
top_k = 6
top_employment = df_clean['Employment'].value_counts().head(top_k).index
df_clean['Employment_group'] = np.where(
    df_clean['Employment'].isin(top_employment), df_clean['Employment'], 'Other'
)

# Order categories by frequency
order = df_clean['Employment_group'].value_counts().index
df_clean['Employment_group'] = pd.Categorical(df_clean['Employment_group'], categories=order, ordered=True)
plt.figure(figsize=(15,9))
sns.scatterplot(data=df_clean, x='JobSatPoints_6', y='ConvertedCompYearly', hue='Employment', size='Employment', sizes=(20, 300), alpha=0.5, palette='magma')
plt.xlabel('Job Satisfaction')
plt.ylabel('Compensation')
plt.title('relationship between yearly compensation Compensation and job satisfaction Job Satisfaction')
```

#### 2. Scatter Plot for Work Experience vs. Age Group by Country


Compare work experience (`YearsCodePro`) across different age groups (`Age`) and countries (`Country`). Use colors to represent different countries and markers for age groups.


```python
## Write your code here
# Get top 10 countries by respondent count
top_countries = plot_data1['Country'].value_counts().nlargest(10).index
filtered_df = plot_data1[plot_data1['Country'].isin(top_countries)]

# Create the plot
plt.figure(figsize=(15, 10))
sns.scatterplot(data=filtered_df, x='Age', y='YearsCodePro_num', hue='Country', palette='viridis', s=100)
plt.title('Professional Coding Experience Across Age Groups and Countries', fontsize=16)
plt.xlabel('Age Group', fontsize=14)
plt.ylabel('Years of Professional Coding Experience', fontsize=14)
plt.xticks(rotation=45)
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
```
