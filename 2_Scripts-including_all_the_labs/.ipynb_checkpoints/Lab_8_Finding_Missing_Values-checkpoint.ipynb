### 3. Finding Missing Values
##### Task 2: Identify missing values for all columns.


```python
## Write your code here
missing_df = df.isnull()
missing_df.head(20)
print(df.isnull().sum())
```

##### Task 3: Visualize missing values using a heatmap (Using seaborn library).



```python
## Write your code here
plt.figure(figsize=(15,6))
sns.heatmap(missing_df, cbar=False)
plt.title('Missing Values Heatmap')
```

##### Task 4: Count the number of missing rows for a specific column (e.g., `Employment`).


```python
## Write your code here
df['EdLevel'].isnull().sum()
# print(df.isnull().sum()) # To see the count of missing values in every column
```

### 4. Imputing Missing Values
##### Task 5: Identify the most frequent (majority) value in a specific column (e.g., `Employment`).


```python
## Write your code here
most_ed = df['EdLevel'].value_counts().idxmax()
print(most_ed)
```

##### Task 6: Impute missing values in the `Employment` column with the most frequent value.



```python
## Write your code here
df['EdLevel'].fillna(most_ed, inplace=True)
# again checking for missing values in edlevel
df['EdLevel'].isnull().sum()
```

### 5. Visualizing Imputed Data
##### Task 7: Visualize the distribution of a column after imputation (e.g., `Employment`).


```python
## Write your code here
plt.figure(figsize=(15,6))
sns.countplot(data=df, x='EdLevel', order=df['EdLevel'].value_counts().index)
plt.xticks(rotation=45)
```
