### Task 1: Trends in Compensation Over Age Groups


##### 1. Line Chart of Median `ConvertedCompYearly` by Age Group


- Track how the median yearly compensation (ConvertedCompYearly) changes across different age groups.

- Use a line chart to visualize these trends.


```python
## Write your code here
# Filter out ambiguous age categories for meaningful trend
excluded_categories = ['Prefer not to say', 'Under 18 years old'] 
df_filtered = df[~df['Age'].isin(excluded_categories)] 
# Group by age and calculate median compensation 
line_data1 = df_filtered.groupby('Age')['ConvertedCompYearly'].median().dropna().sort_index()
line_data1
```

```python
line_data1.plot(kind='line', x ='Age', y='ConvertedCompYearly')

plt.title('Median Yearly Compenation Across Different Age Groups')
plt.xlabel('Age Groups')
plt.ylabel('Yearly Compensation')
plt.xticks(rotation=90)
plt.show()
```

##### 2. Line Chart of Median `ConvertedCompYearly` for Ages 25 to 45


For a closer look, plot a line chart focusing on the median compensation for respondents between ages 25 and 45.


```python
## Write your code here
# Filter out ambiguous age categories for meaningful trend
excluded_categories = ['Prefer not to say', 'Under 18 years old'] 
df_filtered = df[~df['Age'].isin(excluded_categories)] 

line_data2 = df_filtered[df_filtered['Age'].isin(['25-34 years old', '35-44 years old'])]


median_comp_by_age = line_data2.groupby('Age')['ConvertedCompYearly'].median().reset_index()


plt.figure(figsize=(10, 6))
plt.plot(median_comp_by_age['Age'], median_comp_by_age['ConvertedCompYearly'], marker='o', color='teal')
plt.title('Median ConvertedCompYearly for Ages 25 to 45')
plt.xlabel('Age')
plt.ylabel('Median Compensation (Yearly)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Task 2: Trends in Job Satisfaction by Experience Level



##### 1. Line Chart of Job Satisfaction (`JobSatPoints_6`) by Experience Level



- Use a column that approximates experience level to analyze how job satisfaction changes with experience.

- If needed, substitute an available experience-related column for `Experience`.


```python
## Write your code here
# Using YearsCodePro as substitue for experience level
line_data3  = df.dropna(subset=['JobSatPoints_6', 'YearsCodePro'])
line_data3 = line_data3.groupby('YearsCodePro')['JobSatPoints_6'].mean().sort_index()
```

```python
line_data3.plot(kind='line', y = 'JobSatPoints_6', x='YearsCodePro', marker='o', figsize=(15, 6))
plt.grid(True)
plt.title('Trend in Job Satisfaction by Experience level')
plt.ylabel('Median Job Satisfaction')
```

### Task 3: Trends in Job Satisfaction and Compensation by Experience


##### 1.Line Chart of Median ConvertedCompYearly Over Experience Level

- This line chart will track how median compensation (`ConvertedCompYearly`) changes with increasing experience.

- Use a column such as `WorkExp` or another relevant experience-related column.


```python
## Write your code here
line_data4  = df.dropna(subset=['ConvertedCompYearly', 'WorkExp'])

# Remove outliers in ConvertedCompYearly using IQR method
Q1 = df['ConvertedCompYearly'].quantile(0.25)
Q3 = df['ConvertedCompYearly'].quantile(0.75)
IQR = Q3 - Q1
filtered_df = line_data4[(line_data4['ConvertedCompYearly'] >= Q1 - 1.5 * IQR) & (line_data4['ConvertedCompYearly'] <= Q3 + 1.5 * IQR)]


comp_wx = filtered_df.groupby('WorkExp')['ConvertedCompYearly'].mean().sort_index()

comp_wx.plot(kind='line', y = 'ConvertedCompYearly', x='WorkExp')

plt.grid(True)
plt.title('Compensation by Work Experience', fontsize=16)
plt.xlabel('Years of Work Experience', fontsize=12)
plt.ylabel('Average Compensation', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
```

##### 2.Line Chart of Job Satisfaction (`JobSatPoints_6`) Across Experience Levels

- Create a line chart to explore trends in job satisfaction (`JobSatPoints_6`) based on experience level.

- This chart will provide insight into how satisfaction correlates with experience over time


```python
## Write your code here
line_data3  = df.dropna(subset=['JobSatPoints_6', 'WorkExp'])
line_data3 = line_data3.groupby('WorkExp')['JobSatPoints_6'].mean().sort_index()
```

```python
line_data3.plot(kind='line', y = 'JobSatPoints_6', x='WorkExp')
plt.grid(True)
plt.title('Trend in Job Satisfaction by Experience level')
plt.ylabel('Median Job Satisfaction')
```
