## Visualizing Data


### Task 1: Visualizing the Distribution of Data


**1. Box Plot of `CompTotal` (Total Compensation)**


Use a box plot to analyze the distribution and outliers in total compensation.


```python
# your code goes here
q1 = """
SELECT CompTotal, COUNT(*) as count 
FROM main 
GROUP BY CompTotal 
ORDER BY CompTotal
"""
df_comp = pd.read_sql_query(q1, conn)
df_comp.dropna(inplace=True)
print(df_comp)

import numpy as np
df_comp['LogCompTotal'] = np.log10(df_comp['CompTotal'])
df_comp.boxplot(column='LogCompTotal')
# Plot
# df_comp.boxplot(column='CompTotal')
```

**2. Box Plot of Age (converted to numeric values)**


Convert the `Age` column into numerical values and visualize the distribution.


```python
# your code goes here
QUERY = """
SELECT Age, COUNT(*) as count 
FROM main 
GROUP BY Age
ORDER BY Age
"""
df_age = pd.read_sql_query(QUERY, conn)

age_map = {
    '18-24 years old':20,
    '25-34 years old':30,
    '35-44 years old':40,
    '45-54 years old':50,
    '55-64 years old':60,
    '65 years or older':70,
    'Prefer not to say':0,
    'Under 18 years old':15
}
df_age['Age_num'] = df_age['Age'].map(age_map)

df_age.boxplot(column='Age_num')
```

### Task 2: Visualizing Relationships in Data


**1. Box Plot of `CompTotal` Grouped by Age Groups:**


Visualize the distribution of compensation across different age groups.


```python
# your code goes here
q1 = """
SELECT CompTotal, Age 
FROM main
"""
df_comp = pd.read_sql_query(q1, conn)
df_comp.dropna(inplace=True)
# print(df_comp)

# Remove wider outliers using IQR method
Q1 = df_comp['CompTotal'].quantile(0.25)
Q3 = df_comp['CompTotal'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df_cleaned = df_comp[(df_comp['CompTotal'] >= lower_bound) & (df_comp['CompTotal'] <= upper_bound)]

df_cleaned.boxplot(column='CompTotal')
# plt.title('')....
```

**2. Box Plot of `CompTotal` Grouped by Job Satisfaction (`JobSatPoints_6`):**


Examine how compensation varies based on job satisfaction levels.


```python
# your code goes here
q2 = """ SELECT CompTotal, JobSatPoints_6 FROM main
"""
df_cj = pd.read_sql_query(q2, conn)
```

```python
# Drop rows with missing values in relevant columns
required_columns = ['CompTotal', 'JobSatPoints_6']
df_clean = df_cj.dropna(subset=required_columns)
df_clean['JobSatPoints_6'].unique()
```

```python
# mapping job satisfaction
def map_sat(score):
    if score < 30:
        return 'Low'
    elif score < 60:
        return 'Medium'
    elif score < 80:
        return 'High'
    else:
        return 'Very High'

df_clean['JobSat_lvl'] = df_clean['JobSatPoints_6'].apply(map_sat)
print(df_clean['JobSat_lvl'].value_counts())
```

```python
# Remove outliers using IQR method
Q1 = df_comp['CompTotal'].quantile(0.25)
Q3 = df_comp['CompTotal'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out outliers
df_clean = df_clean[(df_clean['CompTotal'] >= lower_bound) & (df_clean['CompTotal'] <= upper_bound)]
```

```python
# Create boxplot to show compensation distribution across job satisfaction levels
plt.figure(figsize=(10, 6))
sns.boxplot(x='JobSat_lvl', y='CompTotal', data=df_clean, palette='viridis')
plt.title('Compensation vs Job Satisfaction')
plt.xlabel('Job Satisfaction Level')
plt.ylabel('Compensation')
```

### Task 3: Visualizing the Composition of Data


**1. Box Plot of `ConvertedCompYearly` for the Top 5 Developer Types:**


Analyze compensation across the top 5 developer roles.


```python
# your code goes here
q3 = """ SELECT ConvertedCompYearly, DevType FROM main
WHERE DevType IN ( SELECT DevType FROM main GROUP BY DevType ORDER BY COUNT(DevType) DESC LIMIT 5)
"""
df_dev = pd.read_sql_query(q3, conn)
df_dev
```

```python
df_dev['DevType'].unique()
```

```python
# Remove outliers using IQR method
Q1 = df_dev['ConvertedCompYearly'].quantile(0.25)
Q3 = df_dev['ConvertedCompYearly'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out outliers
df_clean = df_dev[(df_dev['ConvertedCompYearly'] >= lower_bound) & (df_dev['ConvertedCompYearly'] <= upper_bound)]


plt.figure(figsize=(15, 6))
sns.boxplot(x='DevType', y='ConvertedCompYearly', data=df_clean, palette='viridis')
plt.title('Compensation vs Developer Type')
plt.xticks(rotation=60)
plt.xlabel('Developer Type')
plt.ylabel('Compensation')
```

**2. Box Plot of `CompTotal` for the Top 5 Countries:**


Analyze compensation across respondents from the top 5 countries.


```python
# your code goes here
q4 = """ SELECT CompTotal, Country FROM main
WHERE Country IN ( SELECT Country FROM main GROUP BY Country ORDER BY COUNT(Country) DESC LIMIT 5)
"""
df_country = pd.read_sql_query(q4, conn)

# Remove outliers using IQR method
Q1 = df_country['CompTotal'].quantile(0.25)
Q3 = df_country['CompTotal'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out outliers
df_clean = df_country[(df_country['CompTotal'] >= lower_bound) & (df_country['CompTotal'] <= upper_bound)]
```

```python
plt.figure(figsize=(15, 6))
sns.boxplot(x='Country', y='CompTotal', data=df_clean, palette='viridis')
plt.title('Compensation vs Countries')
plt.xticks(rotation=45)
plt.xlabel('Countries')
plt.ylabel('Compensation')
```

### Task 4: Visualizing Comparison of Data


**1. Box Plot of CompTotal Across Employment Types:**


Analyze compensation for different employment types.


```python
# your code goes here
q5 = """ SELECT CompTotal, Employment FROM main
"""
df_emp = pd.read_sql_query(q5, conn)

# Remove outliers using IQR method
Q1 = df_emp['CompTotal'].quantile(0.25)
Q3 = df_emp['CompTotal'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out outliers
df_clean = df_emp[(df_emp['CompTotal'] >= lower_bound) & (df_emp['CompTotal'] <= upper_bound)]

# Plot
plt.figure(figsize=(15, 6))
sns.boxplot(x='Employment', y='CompTotal', data=df_clean, palette='viridis')
plt.title('Compensation vs Employment Types')
plt.xticks(rotation=90)
plt.xlabel('Employment Types')
plt.ylabel('Compensation')
```

**2. Box Plot of `YearsCodePro` by Job Satisfaction (`JobSatPoints_6`):**


Examine the distribution of professional coding years by job satisfaction levels.


```python
# your code goes here
q5 = """ SELECT YearsCodePro, JobSatPoints_6 FROM main
"""
df_yrs = pd.read_sql_query(q5, conn)

# mapping job satisfaction
def map_sat(score):
    if score < 30:
        return 'Low'
    elif score < 60:
        return 'Medium'
    elif score < 80:
        return 'High'
    else:
        return 'Very High'

df_yrs['JobSat_lvl'] = df_yrs['JobSatPoints_6'].apply(map_sat)

# Clean the 'YearsCodePro' column
df_yrs['YearsCodePro'] = df_yrs['YearsCodePro'].replace('Less than 1 year', '0.5')
df_yrs['YearsCodePro'] = df_yrs['YearsCodePro'].replace('More than 50 years', '50')
df_yrs['YearsCodePro'] = pd.to_numeric(df_yrs['YearsCodePro'], errors='coerce')

# Drop any rows where 'YearsCodePro' became NaN after conversion
df_yrs.dropna(subset=['YearsCodePro'], inplace=True)

# Plot
plt.figure(figsize=(15, 10))
sns.boxplot(y='YearsCodePro', x='JobSat_lvl', data=df_yrs, palette='viridis')
plt.title('years in coding vs JobSat Level')
plt.xticks(rotation=90)
plt.xlabel('JobSat Level')
plt.ylabel('years in coding')
```
