### Task 1: Visualizing Data Composition with Pie Charts


##### 1.1 Create a Pie Chart of the Top 5 Databases Respondents Want to Work With


In the survey data, the `DatabaseWantToWorkWith` column lists the databases that respondents wish to work with. Let’s visualize the top 5 most-desired databases in a pie chart.



```python
##Write your code here
# Strip multiple entries
nomiss = df['DatabaseWantToWorkWith'].dropna()
```

```python
df['DatabaseWantToWorkWith'].unique()
```

```python
unq_db = nomiss.str.split(';').explode().str.strip()
db_counts = unq_db.value_counts()
```

```python
top_des_db = db_counts.head(5)
top_des_db
```

```python
plt.figure(figsize=(8, 8))

colors = plt.cm.Set3(range(len(top_des_db)))
plt.pie(top_des_db, labels=top_des_db.index, autopct='%1.1f%%', startangle=140, colors=colors )
plt.title('Top 5 Databases Respondents Want to Work With')
```

The `DevType` column lists the developer types for respondents. We’ll examine the distribution by showing the top 5 developer roles in a pie chart.



```python
##Write your code here
df['DevType'].unique()
```

```python
# Since, one respondent contains multiple dev roles
# Drop NaNs and split roles
dev_roles = df['DevType'].dropna().str.split(',')
all_roles = dev_roles.explode()
dev_counts = all_roles.value_counts()
top5_dev = dev_counts.head(5)
```

```python
top5_dev
```

```python
plt.pie(top5_dev, labels=top5_dev.index, autopct='%1.1f%%')
plt.title('Top 5 Developer roles')
plt.axis('equal')
```
